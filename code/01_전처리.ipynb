````python
import pandas as pd
import requests
import zipfile
from io import BytesIO
from tqdm import tqdm
import xml.etree.ElementTree as ET
from dotenv import load_dotenv
import os
from collections import Counter
import re

# =====================================================
# 0. OpenDART API KEY 불러오기
# =====================================================

load_dotenv()

API_KEY = (
    os.getenv("OPENDART_API_KEY")
    or os.getenv("DART_API_KEY")
)
if not API_KEY:
    raise ValueError("OpenDART API KEY가 설정되어 있지 않습니다. .env 파일을 확인해주세요.")
````

## 1.1 OpenDART API 사업보고서 수집

**왜**: DART 전자공시 원문 ZIP을 직접 수집해야 사업보고서 섹션별 텍스트를 추출할 수 있다.  
**방법**: `corp_code` → `rcept_no` → 원문 ZIP 다운로드; 회계연도(12월 결산) 기준.  
**결정사항**: 375개 기업 중 ZIP 없는 7개(1.9%) 제외, 3월·6월 결산 기업(신영증권·만호제강) 제외 → 최종 **381개 기업-연도** 수집 시도.

````python
# =====================================================
# 1. 회사 파일 불러오기
# =====================================================

df = pd.read_csv(
    "company_master.csv",
    dtype={"stock_code": str}
 )

df["stock_code"] = (
    df["stock_code"]
    .astype(str)
    .str.zfill(6)
 )

df = df[
    df["fiscal_year"].isin([2022, 2023, 2024])
].copy()

print(" 대상 기업 수:", len(df))
print(df["fiscal_year"].value_counts().sort_index())
````

````python
# =====================================================
# 2. corp_code 테이블 가져오기
# =====================================================

def get_corp_code_table():

    url = (
        "https://opendart.fss.or.kr/api/"
        "corpCode.xml"
    )

    params = {
        "crtfc_key": API_KEY
    }

    res = requests.get(
        url,
        params=params,
        timeout=30
    )

    res.raise_for_status()

    zf = zipfile.ZipFile(
        BytesIO(res.content)
    )

    xml_name = zf.namelist()[0]

    xml_data = zf.read(xml_name)

    root = ET.fromstring(xml_data)

    rows = []

    for item in root.findall("list"):

        stock_code = (
            item.findtext("stock_code")
            or ""
        ).strip()

        if stock_code:

            rows.append({
                "stock_code":
                    stock_code.zfill(6),
                "corp_code":
                    item.findtext(
                        "corp_code"
                    )
            })

    return pd.DataFrame(rows)


print("corp_code 테이블 다운로드 중...")

corp_df = get_corp_code_table()

print(
    "corp_code 개수:",
    len(corp_df)
)
````

````python
# =====================================================
# 3. corp_code merge
# =====================================================

# 기존 corp_code 있으면 제거
if "corp_code" in df.columns:
    df = df.drop(
        columns=["corp_code"]
    )

# stock_code 형식 맞추기
df["stock_code"] = (
    df["stock_code"]
    .astype(str)
    .str.zfill(6)
)

corp_df["stock_code"] = (
    corp_df["stock_code"]
    .astype(str)
    .str.zfill(6)
)

# merge
df = df.merge(
    corp_df[
        [
            "stock_code",
            "corp_code"
        ]
    ],
    on="stock_code",
    how="left"
)

print(
    "corp_code 없는 기업 수:",
    df["corp_code"]
    .isna()
    .sum()
)

print(
    "현재 기업 수:",
    len(df)
)
````

````python
# =====================================================
# 사업보고서(rcept_no) 수집 함수
# - 회계연도 기준 사업보고서 탐색
# - 정정 공시 포함 시 최신본 우선
# =====================================================

def get_report_info(
    corp_code,
    fiscal_year
):

    target_year = (
        int(fiscal_year) + 1
    )

    try:

        all_reports = []

        # 최대 10페이지 순회
        for page_no in range(1, 11):

            params = {
                "crtfc_key":
                    API_KEY,
                "corp_code":
                    str(corp_code)
                    .zfill(8),
                "bgn_de":
                    f"{target_year}0101",
                "end_de":
                    f"{target_year}1231",
                "last_reprt_at":
                    "N",  # 정정본 포함
                "pblntf_detail_ty":
                    "A001",  # 정기공시
                "page_count":
                    100,
                "page_no":
                    page_no
            }

            res = requests.get(
                "https://opendart.fss.or.kr/api/list.json",
                params=params,
                timeout=20
            )

            payload = res.json()

            # API 오류
            if payload.get(
                "status"
            ) != "000":
                break

            page_reports = payload.get(
                "list",
                []
            )

            # 더 이상 데이터 없음
            if len(page_reports) == 0:
                break

            all_reports.extend(
                page_reports
            )

        # 결과 없음
        if len(all_reports) == 0:

            return {
                "status":
                    "not_found",
                "report_nm":
                    None,
                "rcept_no":
                    None,
                "rcept_dt":
                    None,
                "viewer_url":
                    None
            }

        reports = pd.DataFrame(
            all_reports
        )

        # =====================================
        # 해당 연도 사업보고서 필터링
        # =====================================
        target = reports[
            reports["report_nm"]
            .str.contains(
                "사업보고서",
                na=False
            )
        ]

        target = target[
            target["report_nm"]
            .str.contains(
                f"({fiscal_year}.12)",
                regex=False,
                na=False
            )
        ]

        # 못 찾은 경우
        if len(target) == 0:

            return {
                "status":
                    "not_found",
                "report_nm":
                    None,
                "rcept_no":
                    None,
                "rcept_dt":
                    None,
                "viewer_url":
                    None
            }

        # 최신 정정본 우선
        target = target.sort_values(
            "rcept_dt",
            ascending=False
        )

        report = target.iloc[0]

        rcept_no = report[
            "rcept_no"
        ]

        viewer_url = (
            "https://dart.fss.or.kr/"
            "dsaf001/main.do"
            f"?rcpNo={rcept_no}"
        )

        return {
            "status":
                "success",
            "report_nm":
                report[
                    "report_nm"
                ],
            "rcept_no":
                rcept_no,
            "rcept_dt":
                report[
                    "rcept_dt"
                ],
            "viewer_url":
                viewer_url
        }

    except Exception as e:

        print(
            f"ERROR | "
            f"{corp_code} | "
            f"{fiscal_year} | "
            f"{e}"
        )

        return {
            "status":
                "error",
            "report_nm":
                None,
            "rcept_no":
                None,
            "rcept_dt":
                None,
            "viewer_url":
                None
        }
````

````python
results = []

for _, row in tqdm(
    df.iterrows(),
    total=len(df)
):

    report = get_report_info(
        row["corp_code"],
        row["fiscal_year"]
    )

    results.append({
        "company_name":
            row["company_name"],
        "stock_code":
            row["stock_code"],
        "corp_code":
            row["corp_code"],
        "fiscal_year":
            row["fiscal_year"],
        **report
    })

report_df = pd.DataFrame(
    results
)
````

````python
# =====================================================
# 결과 확인
# =====================================================

print("\n수집 결과")

print(
    report_df[
        "status"
    ].value_counts()
)

display(
    report_df.head()
)
# =====================================================
# 실패 기업 확인
# =====================================================

not_found = report_df[
    report_df["status"]
    == "not_found"
].copy()

print(
    "추출 실패 기업 수:",
    len(not_found)
)

display(
    not_found[
        [
            "company_name",
            "stock_code",
            "corp_code",
            "fiscal_year"
        ]
    ]
)
````

#### ❗ 신영증권과 만호제강은 3월·6월 결산 기업
➜ 표본 간 회계연도 기준의 일관성을 확보하기 위해 분석 대상에서 제외

````python
# 제거할 기업
remove_companies = [
    "신영증권",
    "만호제강"
]

final_df = report_df[
    ~report_df["company_name"]
    .isin(remove_companies)
].copy()

# 저장
final_df.to_csv(
    "사업보고서전체.csv",
    index=False,
    encoding="utf-8-sig"
)

print("저장 완료")
print("최종 기업 수:", len(final_df))


# =====================================
# raw_zip 폴더 생성
# =====================================

BASE_DIR = "raw_zip"

YEARS = [
    2022,
    2023,
    2024
]

for year in YEARS:

    os.makedirs(
        os.path.join(
            BASE_DIR,
            str(year)
        ),
        exist_ok=True
    )


# =====================================
# ZIP 다운로드 함수
# =====================================

def download_raw_zip(
    rcept_no,
    fiscal_year
):

    save_path = os.path.join(
        BASE_DIR,
        str(fiscal_year),
        f"{rcept_no}.zip"
    )

    # 이미 존재하면 skip
    if os.path.exists(
        save_path
    ):
        return "already_exists"

    try:

        params = {
            "crtfc_key":
                API_KEY,
            "rcept_no":
                rcept_no
        }

        res = requests.get(
            "https://opendart.fss.or.kr/api/document.xml",
            params=params,
            timeout=60
        )

        # 응답 실패
        if res.status_code != 200:
            return "download_fail"

        # zip 저장
        with open(
            save_path,
            "wb"
        ) as f:

            f.write(
                res.content
            )

        return "success"

    except Exception:

        return "error"


# =====================================
# 전체 기업 ZIP 다운로드
# =====================================

download_results = []

for _, row in tqdm(
    final_df.iterrows(),
    total=len(final_df)
):

    # rcept_no 없는 경우 skip
    if pd.isna(
        row["rcept_no"]
    ):

        status = "no_rcept_no"

    else:

        status = download_raw_zip(
            row["rcept_no"],
            row["fiscal_year"]
        )

    download_results.append(
        status
    )

final_df[
    "zip_download_status"
] = download_results


# =====================================
# 결과 확인
# =====================================

print(
    final_df[
        "zip_download_status"
    ].value_counts()
)
````

````python
# =====================================
# ZIP 유효성 검사
# =====================================

invalid_zip_rows = []

for idx, row in final_df.iterrows():

    rcept_no = str(row["rcept_no"])
    fiscal_year = row["fiscal_year"]

    zip_path = os.path.join(
        BASE_DIR,
        str(fiscal_year),
        f"{rcept_no}.zip"
    )

    # 파일 없는 경우
    if not os.path.exists(zip_path):

        invalid_zip_rows.append(idx)
        continue

    # ZIP 정상 여부 확인
    try:
        with zipfile.ZipFile(
            zip_path,
            "r"
        ) as zf:

            zf.namelist()

    except Exception:

        invalid_zip_rows.append(idx)


# =====================================
# 실패 기업 확인
# =====================================

failed_zip_df = final_df.loc[
    invalid_zip_rows,
    [
        "company_name",
        "corp_code",
        "rcept_no",
        "fiscal_year",
        "report_nm"
    ]
].reset_index(drop=True)

print("❌ ZIP 파일 오류 기업")
print(
    f"{len(failed_zip_df)}개"
)

display(failed_zip_df)


# =====================================
# 실패 기업 제거
# =====================================

final_df_clean = final_df.drop(
    invalid_zip_rows
).reset_index(drop=True)


# =====================================
# CSV 재저장
# =====================================

final_df_clean.to_csv(
    "사업보고서전체.csv",
    index=False,
    encoding="utf-8-sig"
)

print("✅ 저장 완료")
print(
    "최종 기업 수:",
    len(final_df_clean)
)
````

## 2.1 Seed 사전 보완 및 문장 추출

**왜**: 단일 키워드는 다의성이 높아 오탐(False Positive)이 많이 발생한다.  
**방법**: 복합어 확장(안전→안전보건·산업안전), compliance 중복 토큰 제거, FastText 기반 유사어 보완.  
**결정사항**: E/S/G 3차원 Seed 사전 보완 완료 — II/IV/VI 섹션에서 문장 추출.

## 🔧 2-1. compliance 중복 토큰 제거

``` text
 Before:
     준법|compliance
     컴플라이언스|compliance

 After:
     준법|컴플라이언스|compliance
```

➥ 💡 **차원의 seed dictionary에서 동일 영어 토큰이 중복 포함되는 문제를 보완**

````python
# =====================================================
# compliance 중복 제거 후
# 차원별 통합 OR 패턴 생성
# =====================================================

import pandas as pd
import re
seed_df = pd.read_csv(
    "seed_dictionary.csv"
)

DIM_PATTERNS = {}

for dim in ["E", "S", "G"]:

    subset = seed_df[
        seed_df["dimension"] == dim
    ]

    all_tokens = []

    # 모든 pattern 토큰 수집
    for pattern in subset["pattern"]:

        tokens = [
            t.strip()
            for t in str(pattern).split("|")
            if t.strip()
        ]

        all_tokens.extend(tokens)

    # =====================================
    # compliance 중복 제거
    # =====================================

    seen_compliance = False
    cleaned_tokens = []

    for token in all_tokens:

        if token.lower() == "compliance":

            if seen_compliance:
                continue

            seen_compliance = True

        cleaned_tokens.append(token)

    # 최종 OR 패턴
    DIM_PATTERNS[dim] = re.compile(
    "|".join(map(re.escape, cleaned_tokens))
)

print("G pattern:\n")
print(
    DIM_PATTERNS["G"]
)
````

## 🔧 2-3. ESG 문장 추출 함수 정의

#### 🟨 ESG 문장 추출 함수

> 사업보고서 XML 데이터를 기반으로 ESG 문장을 추출하기 위한 핵심 함수 모음

| 단계 | 기능 |
|:---|:---|
| 🟩 XML → Text | XML 데이터를 Plain Text로 변환 |
| 🟩 TOC 제거 | 목차 제거 |
| 🟩 섹션 추출 | **II / IV / VI** 섹션 추출 |
| 🟩 문장 분리 | 텍스트를 문장 단위로 분리 |

````python
# =====================================================
# ESG 문장 추출 함수 모음
# - XML → plain text
# - TOC 제거
# - II / IV / VI 섹션 추출
# - 문장 분리
# =====================================================

# ─────────────────────────────────────────────────────
# 1. XML → plain text
# (2단계 블록태그 보존법)
# ─────────────────────────────────────────────────────

def xml_to_plain(
    zip_path: str,
    rcept_no: str
) -> str:
    """
    ZIP 내 XML을 읽어
    섹션 경계를 보존한 plain text 반환

    1단계:
        블록 태그 → \n

    2단계:
        나머지 태그 → 공백
    """

    with zipfile.ZipFile(zip_path) as zf:

        names = zf.namelist()

        target = (
            f"{rcept_no}.xml"
            if f"{rcept_no}.xml" in names
            else names[0]
        )

        raw = zf.read(target).decode(
            "utf-8",
            errors="ignore"
        )

    BLOCK = (
        r"p|div|tr|br|"
        r"h[1-6]|li|ul|ol|"
        r"table|thead|tbody|section"
    )

    # 블록 태그 개행 처리
    raw = re.sub(
        rf"<(?:{BLOCK})(?:\s[^>]*)?>",
        "\n",
        raw,
        flags=re.IGNORECASE
    )

    raw = re.sub(
        rf"</(?:{BLOCK})>",
        "\n",
        raw,
        flags=re.IGNORECASE
    )

    # 나머지 태그 제거
    text = re.sub(
        r"<[^>]+>",
        " ",
        raw
    )

    # 공백 정리
    text = re.sub(
        r"[ \t]+",
        " ",
        text
    )

    text = re.sub(
        r"\n{3,}",
        "\n\n",
        text
    )

    return text.strip()


# ─────────────────────────────────────────────────────
# 2. TOC 감지
# ─────────────────────────────────────────────────────

def is_toc(text: str) -> bool:
    """
    목차(TOC) 여부 판별
    """

    lines = [
        l.strip()
        for l in text.split("\n")
        if l.strip()
    ]

    if not lines:
        return False

    toc_count = 0

    for l in lines:

        if re.search(r"\.{4,}", l):
            toc_count += 1

        elif re.search(r"-{4,}", l):
            toc_count += 1

        elif re.fullmatch(r"\d+", l):
            toc_count += 1

        elif re.search(
            r"[-\.]{3,}\s*\d+\s*$",
            l
        ):
            toc_count += 1

        elif re.search(
            r"[\.]{6,}\s*\d+\s*$",
            l
        ):
            toc_count += 1

    return (
        toc_count / len(lines)
    ) > 0.25


# ─────────────────────────────────────────────────────
# 3. 섹션 분리
# II / IV / VI
# ─────────────────────────────────────────────────────

SECTION_PATTERNS = {

    "II": (
        r"\n[ \t]*(?:제\s*2\s*장|II|Ⅱ)"
        r"\s*[\s\.][^\n]{0,20}"
        r"사업의[^\n]*내용[^\n]*\n"
        r"(.+?)"
        r"(?="
        r"\n[ \t]*(?:제\s*3\s*장|III|Ⅲ)"
        r"\s*[\s\.][^\n]{0,20}재무"
        r"|"
        r"\n[ \t]*(?:제\s*4\s*장|IV|Ⅳ)"
        r"\s*[\s\.][^\n]{0,30}"
        r"경영진단"
        r"|$)"
    ),

    "IV": (
        r"\n[ \t]*(?:제\s*4\s*장|IV|Ⅳ)"
        r"\s*[\s\.][^\n]{0,30}"
        r"경영진단[^\n]*\n"
        r"(.+?)"
        r"(?="
        r"\n[ \t]*(?:제\s*5\s*장|V|Ⅴ)"
        r"\s*[\s\.][^\n]{0,20}"
        r"회계감사"
        r"|"
        r"\n[ \t]*(?:제\s*6\s*장|VI|Ⅵ)"
        r"\s*[\s\.][^\n]{0,10}"
        r"이사회"
        r"|$)"
    ),

    "VI": (
        r"\n[ \t]*(?:제\s*6\s*장|VI|Ⅵ)"
        r"\s*[\s\.][^\n]{0,10}"
        r"이사회[^\n]*\n"
        r"(.+?)"
        r"(?="
        r"\n[ \t]*(?:제\s*7\s*장|VII|Ⅶ)"
        r"\s*[\s\.][^\n]{0,10}"
        r"주주"
        r"|"
        r"\n[ \t]*(?:제\s*8\s*장|VIII|Ⅷ)"
        r"\s*[\s\.][^\n]{0,10}"
        r"임원"
        r"|$)"
    ),
}


def extract_sections(
    plain_text: str
) -> dict:
    """
    II / IV / VI
    섹션 추출
    """

    result = {}

    for key, pat in (
        SECTION_PATTERNS.items()
    ):

        found = ""

        for m in re.finditer(
            pat,
            plain_text,
            re.DOTALL | re.IGNORECASE
        ):

            candidate = (
                m.group(1)
                .strip()
            )

            if not is_toc(
                candidate
            ):
                found = candidate
                break

        result[key] = found

    return result


# ─────────────────────────────────────────────────────
# 4. 문장 분리
# ─────────────────────────────────────────────────────

def split_sentences(
    text: str
) -> list[str]:

    pattern = (
        r"(?<=[^0-9])"
        r"(?<=[.?!])\s+"
    )

    sents = re.split(
        pattern,
        text.strip()
    )

    return [
        s.strip()
        for s in sents
        if s.strip()
    ]

````

````python
import pandas as pd
from pathlib import Path
from tqdm import tqdm
import os


RAW_DIR = Path("raw_zip")

report_df = pd.read_csv(
    "사업보고서전체.csv",
    dtype={
        "stock_code": str,
        "corp_code": str,
        "rcept_no": str
    }
)

report_df["stock_code"] = (
    report_df["stock_code"]
    .str.zfill(6)
)

report_df["corp_code"] = (
    report_df["corp_code"]
    .str.zfill(8)
)

report_df["rcept_no"] = (
    report_df["rcept_no"]
    .astype(str)
)

# 결과 저장용
all_rows = []

for _, row in tqdm(
    report_df.iterrows(),
    total=len(report_df)
):

    rcept_no = str(
        row["rcept_no"]
    )

    year = str(
        row["fiscal_year"]
    )

    zip_path = os.path.join(
        RAW_DIR,
        year,
        f"{rcept_no}.zip"
    )

    if not os.path.exists(
        zip_path
    ):
        continue

    try:

        plain_text = xml_to_plain(
            zip_path,
            rcept_no
        )

        sections = extract_sections(
            plain_text
        )

        for section, text in (
            sections.items()
        ):

            if not text.strip():
                continue

            sentences = split_sentences(
                text
            )

            for sent in sentences:

                sent = sent.strip()

                if len(sent) < 20:
                    continue

                if is_toc(sent):
                    continue

                all_rows.append({

                    "stock_code":
                        row["stock_code"],

                    "fiscal_year":
                        row["fiscal_year"],

                    "section":
                        section,

                    "sentence":
                        sent
                })

    except Exception as e:

        print(
            rcept_no,
            e
        )
        

sentence_df = pd.DataFrame(
    all_rows
)

print(
    "전체 문장 수:",
    len(sentence_df)
)

print(
    "섹션별 문장 수:"
)

print(
    sentence_df[
        "section"
    ].value_counts()
)

display(
    sentence_df.head(20)
)
````

> **📌 STRUCT_NOISE 제거 기준**  

> 구조 노이즈는 전체 섹션(**II / IV / VI**)에서 탐색
>> **법정 공시 형식의 반복 문장 패턴은 VI 섹션(이사회·거버넌스)에 집중적으로 분포**하는 것으로 확인

| 구분 | II / IV 섹션 | VI 섹션 |
|:---|:---|:---|
| 📄 섹션 성격 | 사업 내용, 경영진단 및 분석 | 이사회·위원회·지배구조 공시 |
| 🔢 숫자 패턴 의미 | **ESG 성과·정책·리스크 정보와 결합 가능** | **법정 기재 중심 절차 정보 비중 높음** |
| 📝 예시 | 탄소배출 **N% 감소**, 안전교육 **N회 실시**, ESG 목표 **N년 계획** | 이사 **N인 구성**, 임기 **N년**, **YYYY.MM.DD 선임** |
| ⚙️ 처리 방식 | **유지** | **제거** |

### 🔍 VI 섹션 주요 제거 패턴

| 유형 | 예시 |
|:---|:---|
| 👥 이사회 구성 | 이사 **N인** 구성 |
| 📅 임기·선임 | 임기 **N년**, **YYYY.MM.DD 선임** |
| 🔄 인사 변동 | 재선임 / 사임 |
| 🗳️ 회의 정보 | 출석률, 중요의결사항 |
| 🏛️ 위원회 | 구성·현황·개최 내역 |


> ❌ **VI 섹션의 반복 문장은 기업 간 차별적 ESG 의미보다 법정 기재·절차 정보 성격이 강하다고 판단하여 제거**

````python
# =============================================
# STRUCTURE_NOISE 패턴 정의
# =============================================
STRUCT_NOISE = re.compile(
    r"이사\s*\d+\s*인|"
    r"임기\s*\d+\s*년|"
    r"\d{4}\.\d{1,2}\.\d{1,2}\s*선임|"
    r"재선임|"
    r"사임|"
    r"출석(?:률|하였|한)|"
    r"중요의결사항|"
    r"위원회\s*구성|"
    r"위원회\s*현황|"
    r"개최\s*현황",
    re.IGNORECASE
)
````

1차 ESG 문장 추출

````python
# ─────────────────────────────────────────────────────
# ESG passage 추출 
# ─────────────────────────────────────────────────────

def extract_passages(
    text: str,
    section_label: str,
    meta: dict,
    seen_sentences: set,
    min_len: int = 20,
    max_len: int = 600,
):

    # 0) 목차 자체 skip
    if is_toc(text):
        return []

    rows = []

    sentences = split_sentences(text)

    for sent in sentences:

        sent = sent.strip()

        # 1) 길이 필터
        if not (min_len <= len(sent) <= max_len):
            continue

        # 2) TOC 제거
        if is_toc(sent):
            continue

        # 3) 표 구조 제거
        if sent.count("\n") >= 3:
            continue

        # 4) boilerplate 제거
        if BOILERPLATE.search(sent):
            continue

        # 5) structure noise 제거
        if STRUCT_NOISE.search(sent):
            continue

        # 6) 중복 제거 (기업 + 연도 + 문장 기준)
        sent_key = (
            meta["stock_code"],
            meta["fiscal_year"],
            sent
        )

        if sent_key in seen_sentences:
            continue

        matched_any = False

        # ─────────────────────────────
        # ESG 매칭 (E / S / G)
        # ─────────────────────────────
        for dim, pat in DIM_PATTERNS.items():

            matches = re.findall(pat, sent, flags=re.IGNORECASE)

            if not matches:
                continue

            matched_any = True

            # ─────────────────────────────
            # flatten 안전 처리 (핵심 수정)
            # ─────────────────────────────
            flat_matches = []

            for m in matches:

                if isinstance(m, tuple):
                    flat_matches.extend([x for x in m if x])
                else:
                    flat_matches.append(m)

            # 순서 유지 + 중복 제거
            matched_toks = list(dict.fromkeys(flat_matches))

            # seed term 변환 (사람이 읽는 ESG 개념)
            seed_terms = list(
                dict.fromkeys(
                    SEED_MAP.get(tok, (dim, tok))[1]
                    for tok in matched_toks
                )
            )

            rows.append({
                **meta,
                "section": section_label,
                "dimension": dim,
                "seed_term": "|".join(seed_terms),
                "matched_text": "|".join(matched_toks),
                "matched_sentence": sent,
            })

        # ESG 문장만 seen 처리
        if matched_any:
            seen_sentences.add(sent_key)

    return rows

# =====================================================
# ESG passage 추출 
# =====================================================

import os
from tqdm import tqdm

# 중복 제거용 set
seen_sentences = set()

# 결과 저장용 리스트
esg_rows = []

# 통계
error_count = 0
success_count = 0

# 모든 보고서 순회
for _, row in tqdm(
    report_df.iterrows(),
    total=len(report_df),
    desc="ESG passage 추출"
):

    stock_code = row["stock_code"]
    fiscal_year = row["fiscal_year"]
    rcept_no = str(row["rcept_no"])

    zip_path = os.path.join(
        RAW_DIR,
        str(fiscal_year),
        f"{rcept_no}.zip"
    )

    if not os.path.exists(zip_path):
        continue

    try:

        # 1) XML → plain text
        plain_text = xml_to_plain(
            zip_path,
            rcept_no
        )

        # 2) 섹션 추출 (II, IV, VI)
        sections = extract_sections(
            plain_text
        )

        # 3) 메타 정보
        meta = {
            "stock_code": stock_code,
            "fiscal_year": fiscal_year,
            "rcept_no": rcept_no,
        }

        # 4) 섹션별 ESG 문장 추출
        for section_label, section_text in (
            sections.items()
        ):

            if not section_text.strip():
                continue

            passages = extract_passages(
                section_text,
                section_label,
                meta,
                seen_sentences
            )

            esg_rows.extend(passages)

        success_count += 1

    except Exception as e:

        error_count += 1

        if error_count <= 5:
            print(
                f"❌ {rcept_no}: {str(e)[:60]}"
            )

# =====================================================
# 결과 데이터프레임 생성
# =====================================================

esg_df = pd.DataFrame(esg_rows)

print(f"\n✅ ESG passage 추출 완료")
print(f"   - 처리된 보고서: {success_count}건")
print(f"   - 추출된 ESG 문장: {len(esg_df)}건")
print(f"   - 오류 발생: {error_count}건")
print()
print("차원별 분포:")
print(esg_df["dimension"].value_counts())
print()
print("섹션별 분포:")
print(esg_df["section"].value_counts())
print()
print("샘플 데이터:")
display(esg_df.head(10))
````

### ✔️ 2-4-3. 🧩 G Seed 재구성

#### ☑️ G 단일 축이 아닌 **Core / Structure / G-S Hybrid**로 구분

| 유형 | 설명 | 예시 Seed |
|:---|:---|:---|
| 🟢 **G-core** | 지배구조·권한 견제·감시 기능 | 주주권, 독립성, 의결권 |
| 🟡 **G-structure** | 법정 공시 형식의 구조·현황 정보 | 이사회, 사외이사, 위원회 |
| 🟠 **G-S Hybrid** | 행동 규범·준법·통제 영역 | 준법, 윤리경영, 부패방지 |

> ⚠️ **기존 G 과다 문제는 `G-core`와 `G-structure`가 혼합되어 발생한 것으로 판단**

---

#### ☑️ 준법(Compliance) Seed의 위치

##### 준법 Seed는 **G와 S 사이의 Hybrid 영역**

| 관점 | 설명 |
|:---|:---|
| 🟢 **G-core로 보기 어려운 이유** | 준법은 **지배구조 자체보다 행동·규범·통제 성격**이 강함 |
| 🔵 **S와 연결되는 이유** | 윤리경영, 부패방지, 내부 규정 준수 등 **사회적 책임 요소 포함** |

##### 🎯 대응 방식

```python
G_S_SEEDS = {
    "준법",
    "컴플라이언스",
    "부패방지",
    "윤리",
    "윤리경영"
}
```

> **준법(Compliance)은 G와 S 특성을 동시에 가지므로 `G-S Hybrid` 축으로 별도 관리**

---

#### ☑️ 감사위원회 Seed의 두 얼굴

##### 동일 Seed라도 **문맥(context)에 따라 역할이 달라짐**

| 구분 | 설명 | 예시 |
|:---|:---|:---|
| 🟢 **G-core** | 감시·견제 기능 수행 | 경영 감시, 내부통제, 주주 보호 |
| 🟡 **G-structure** | 공시 구조·현황 설명 | 감사위원회 구성, 운영 현황, 설치 여부 |

> 🔍 **동일한 ‘감사위원회’라도 문맥에 따라 `G-core` 또는 `G-structure`로 구분 가능**

- 🟢 **무엇을 했는지(기능·활동)** → `G-core`  
- 🟡 **구성·설치·현황 설명** → `G-structure`

````python
seed_df = pd.read_csv(
    "seed_dictionary.csv"
)

# =====================================
# G Seed 체계 정의
# =====================================

# 1. Governance Core
G_CORE_BASE = {
    "주주",
    "독립성",
    "의결권"
}

# 2. Governance-Social Hybrid
G_S_SEEDS = {
    "준법",
    "컴플라이언스",
    "부패방지",
    "윤리",
    "윤리경영"
}

# 3. 문맥 따라 달라지는 seed
G_AMBIGUOUS = {
    "이사회",
    "사외이사",
    "감사위원회"
}

# 4. 구조 정보 전용 seed
G_STRUCTURE_ONLY = {
    "위원회"
}

# =====================================
# Context Hint
# =====================================

STRUCTURE_HINTS = {
    "구성",
    "현황",
    "설치",
    "운영",
    "개최",
    "명단",
    "보고",
    "현황보고",
    "선임",
    "재선임"
}

CORE_HINTS = {
    "감시",
    "통제",
    "견제",
    "권한",
    "의결",
    "보호",
    "책임"
}

# =====================================
# Seed Group Mapping
# (Exploratory Analysis 용)
# =====================================

def seed_group(seed_term):

    if seed_term in G_CORE_BASE:
        return "G_CORE"

    if seed_term in G_S_SEEDS:
        return "G_S"

    if seed_term in G_AMBIGUOUS:
        return "G_AMBIGUOUS"

    if seed_term in G_STRUCTURE_ONLY:
        return "G_STRUCTURE"

    return "OTHER"


seed_df["seed_group"] = (
    seed_df["seed_term"]
    .apply(seed_group)
)

# =====================================
# Final G Classification
# =====================================

def classify_final_g(
    seed_term,
    sentence
):

    sent = sentence

    # ----------------------
    # 1. 고정 G-core
    # ----------------------
    if seed_term in G_CORE_BASE:
        return "G_CORE"

    # ----------------------
    # 2. G-S Hybrid
    # ----------------------
    if seed_term in G_S_SEEDS:
        return "G_S"

    # ----------------------
    # 3. Ambiguous Seed
    # ----------------------
    if seed_term in G_AMBIGUOUS:

        structure_score = sum(
            hint in sent
            for hint in STRUCTURE_HINTS
        )

        core_score = sum(
            hint in sent
            for hint in CORE_HINTS
        )

        if structure_score > core_score:
            return "G_STRUCTURE"

        if core_score > structure_score:
            return "G_CORE"

        # default:
        # governance reporting bias 고려
        return "G_STRUCTURE"

    # ----------------------
    # 4. Pure Structure
    # ----------------------
    if seed_term in G_STRUCTURE_ONLY:
        return "G_STRUCTURE"

    return "OTHER"


# =====================================
# ESG Passage Extraction
# =====================================

def extract_passages(
    text: str,
    section_label: str,
    meta: dict,
    seen_sentences: set,
    min_len: int = 20,
    max_len: int = 600,
):

    if is_toc(text):
        return []

    rows = []
    sentences = split_sentences(text)

    for sent in sentences:

        sent = sent.strip()

        # =================================
        # 1. Basic Filters
        # =================================
        if not (
            min_len
            <= len(sent)
            <= max_len
        ):
            continue

        if is_toc(sent):
            continue

        if sent.count("\n") >= 3:
            continue

        if BOILERPLATE.search(sent):
            continue

        if STRUCT_NOISE.search(sent):
            continue

        # =================================
        # 2. Duplicate Check
        # =================================
        sent_key = (
            meta["stock_code"],
            meta["fiscal_year"],
            sent
        )

        if sent_key in seen_sentences:
            continue

        matched_any = False

        # =================================
        # 3. ESG Matching
        # =================================
        for dim, pat in DIM_PATTERNS.items():

            matches = re.findall(
                pat,
                sent,
                flags=re.IGNORECASE
            )

            if not matches:
                continue

            matched_any = True

            # flatten
            flat_matches = []

            for m in matches:

                if isinstance(
                    m,
                    tuple
                ):
                    flat_matches.extend(
                        [
                            x for x in m
                            if x
                        ]
                    )
                else:
                    flat_matches.append(m)

            matched_toks = list(
                dict.fromkeys(
                    flat_matches
                )
            )

            # =================================
            # 4. Seed Mapping
            # =================================
            seed_terms_raw = list(
                dict.fromkeys(
                    SEED_MAP.get(
                        tok,
                        (dim, tok)
                    )[1]
                    for tok in matched_toks
                )
            )

            # =================================
            # 5. G Type Classification
            # =================================
            g_types = list(
                dict.fromkeys(
                    classify_final_g(
                        seed,
                        sent
                    )
                    for seed in seed_terms_raw
                )
            )

            # =================================
            # 6. Structure Filter
            # STRUCTURE 단독 제거
            # =================================
            is_all_structure = all(
                gt == "G_STRUCTURE"
                for gt in g_types
            )

            is_structure_dominant = (
                ("G_STRUCTURE" in g_types)
                and ("G_CORE" not in g_types)
                and ("G_S" not in g_types)
            )

            if (
                is_all_structure
                or is_structure_dominant
            ):
                continue

            # =================================
            # 7. Save Row
            # =================================
            rows.append({
                **meta,
                "section":
                    section_label,
                "dimension":
                    dim,

                "seed_term":
                    "|".join(
                        seed_terms_raw
                    ),

                "G_type":
                    "|".join(
                        g_types
                    ),

                "matched_text":
                    "|".join(
                        matched_toks
                    ),

                "matched_sentence":
                    sent,
            })

        # =================================
        # 8. Seen Update
        # =================================
        if matched_any:
            seen_sentences.add(
                sent_key
            )

    return rows
````

````python
# =====================================
# 초기화
# =====================================

seen_sentences = set()
esg_rows = []

error_count = 0
success_count = 0


# =====================================
# 전체 보고서 순회
# =====================================

for _, row in tqdm(
    report_df.iterrows(),
    total=len(report_df),
    desc="ESG 재추출 (G-core/structure 반영)"
):

    stock_code = row["stock_code"]
    fiscal_year = row["fiscal_year"]
    rcept_no = str(row["rcept_no"])

    zip_path = os.path.join(
        RAW_DIR,
        str(fiscal_year),
        f"{rcept_no}.zip"
    )

    # ZIP 없는 경우 skip
    if not os.path.exists(zip_path):
        continue

    try:

        # =====================================
        # 1. XML → Plain Text
        # =====================================
        plain_text = xml_to_plain(
            zip_path,
            rcept_no
        )

        # =====================================
        # 2. Section Split
        # =====================================
        sections = extract_sections(
            plain_text
        )

        meta = {
            "stock_code":
                stock_code,
            "fiscal_year":
                fiscal_year,
            "rcept_no":
                rcept_no,
        }

        # =====================================
        # 3. Section Loop
        # =====================================
        for (
            section_label,
            section_text
        ) in sections.items():

            if (
                not section_text
                or not section_text.strip()
            ):
                continue

            # =====================================
            # 4. ESG Passage Extraction
            # =====================================
            rows = extract_passages(
                text=section_text,
                section_label=section_label,
                meta=meta,
                seen_sentences=seen_sentences
            )

            esg_rows.extend(rows)

        success_count += 1

    except Exception as e:

        error_count += 1

        # 오류 샘플 일부만 출력
        if error_count <= 5:
            print(
                f"❌ {rcept_no}: "
                f"{str(e)[:80]}"
            )


# =====================================
# 결과 DataFrame 생성
# =====================================

esg_df = pd.DataFrame(
    esg_rows
)


# =====================================
# 결과 확인
# =====================================

print("\n✅ ESG 재추출 완료")
print(f"   - 성공: {success_count}")
print(f"   - 오류: {error_count}")
print(f"   - 총 ESG 문장: {len(esg_df)}")


if "dimension" in esg_df.columns:
    print("\ndimension 분포:")
    print(
        esg_df["dimension"]
        .value_counts()
    )


if "G_type" in esg_df.columns:
    print("\nG_type 분포:")
    print(
        esg_df["G_type"]
        .value_counts()
    )


if "section" in esg_df.columns:
    print("\nsection 분포:")
    print(
        esg_df["section"]
        .value_counts()
    )


display(
    esg_df.head(10)
)


# =====================================
# Export (팀 공유용)
# =====================================

df_export = esg_df.copy()

# G_type 제거
if "G_type" in df_export.columns:
    df_export = df_export.drop(
        columns=["G_type"]
    )

# 컬럼 순서 정리
cols = [
    "stock_code",
    "fiscal_year",
    "rcept_no",
    "section",
    "dimension",
    "seed_term",
    "matched_text",
    "matched_sentence"
]

existing_cols = [
    c for c in cols
    if c in df_export.columns
]

df_export = df_export[
    existing_cols
]

# 저장
output_path = (
    "ESG_passages_clean.csv"
)

df_export.to_csv(
    output_path,
    index=False,
    encoding="utf-8-sig"
)

print(
    f"\n✅ 저장 완료: "
    f"{output_path}"
)
print(
    f"shape: "
    f"{df_export.shape}"
)
````

| `정관\s*제\d+조` | 동일 |

→ 단독으로 제거하면 실제 G 관련 내용까지 날릴 위험 → **Claude API 분류(C단계)로 위임**

````python
df = pd.read_csv("ESG_passages_clean.csv")
before_a2 = len(df)

# ─── A-2 패턴 정의 ────────────────────────────────────────────────────
PATTERN_A2_SAFE = (
    r'발행주식총수의\s*100분의'
    r'|신주를\s*배정하는\s*경우에는\s*주주에게\s*신주인수권'
)

PATTERN_A2_RISKY = (
    r'상법\s*제\d+조'
    r'|제\d+항\s*각\s*호'
    r'|제\d+조\s*제\d+항'
    r'|정관\s*제\d+조'
)

mask_safe  = df['matched_sentence'].str.contains(PATTERN_A2_SAFE,  regex=True, na=False)
mask_risky = df['matched_sentence'].str.contains(PATTERN_A2_RISKY, regex=True, na=False)

print("─── A-2 안전 패턴 제거 대상 샘플 (최대 5건) ────────────────────────────")
for _, row in df[mask_safe].head(5).iterrows():
    print(f"  seed={row['seed_term']} | dim={row['dimension']} | sec={row['section']}")
    print(f"  {row['matched_sentence'][:180]}")
    print()

print("─── A-2 C단계 플래그 대상 샘플 (최대 5건) ──────────────────────────────")
for _, row in df[~mask_safe & mask_risky].head(5).iterrows():
    print(f"  seed={row['seed_term']} | dim={row['dimension']} | sec={row['section']}")
    print(f"  {row['matched_sentence'][:180]}")
    print()

# ─── C단계 플래그 추가 (제거 전에 먼저 설정) ───────────────────────────
if 'c_stage_flag' not in df.columns:
    df['c_stage_flag'] = False
df.loc[~mask_safe & mask_risky, 'c_stage_flag'] = True

c_flagged = (~mask_safe & mask_risky).sum()
print(f"C단계 플래그 추가: {c_flagged}건")

# ─── 안전 패턴 제거 실행 ─────────────────────────────────────────────
removed_a2 = mask_safe.sum()
df = df[~mask_safe].reset_index(drop=True)
after_a2 = len(df)

print(f"A-2 제거: {removed_a2}건")
print(f"  전: {before_a2:,}  →  후: {after_a2:,}")
print()
print("dimension 분포 (A-2 후):")
print(df['dimension'].value_counts())
print()
print(f"C단계 플래그 잔존: {df['c_stage_flag'].sum()}건")

df.to_csv("ESG_passages_clean.csv", index=False)
````

## 🗑️ A-3. 임원선임 절차·주총 결의 기록 제거 + 작성기준일 C단계 플래그

### 발견 경위

## 🔍 B단계. dimension별 직접 읽기 → 추가 패턴 발견

E / S / G 각 30개씩 무작위 샘플을 육안으로 확인한 결과입니다.

### 발견된 추가 오탐 패턴

#### ✅ 추가 제거 (B-1 ~ B-5)

| | **합계** | **339** | |

#### ⚠️ C단계 플래그 추가 (139건)
| 패턴 | 건수 | 판단 근거 |
|:---|:---|:---|
| `임원 및 직원 등에 관한 사항` (>100자) | 4 | 이사 선임 절차 내용 포함 가능 |
| `삼성SDI\|LG에너지솔루션\|SK온` | 33 | 배터리 업계 산업구조 서술 — ESG 아닐 가능성 높으나 문맥 의존 |

````python
df = pd.read_csv("ESG_passages_clean.csv")
before_b = len(df)

# ─── B-1: 신주인수권 계열 공백 보완 ──────────────────────────────────
P1 = (
    r'발행주식\s*총수의\s*100분의'
    r'|신주인수의\s*청약을\s*할\s*기회를\s*부여'
    r'|신주\s*인수권\s*증서'
    r'|주주외의\s*자에게\s*신주를\s*배정'
    r'|신주를\s*배정하는\s*경우에는'
)
# ─── B-2: 전력을 다하 (全力 관용어) ─────────────────────────────────
P2 = r'전력을\s*다하'

# ─── B-3: 노동집약·노동자본 (생산요소) ──────────────────────────────
P3 = r'노동집약|노동[\s·]+자본'

# ─── B-4: 임원 및 직원 목차 참조 ─────────────────────────────────────
#   ≤100자 → 제거 | >100자(이사 선임 내용 포함) → C플래그 | ESG위원회 포함 → 보존
P4 = r'임원\s*(및|과)\s*직원\s*등에\s*관한\s*사항'

# ─── B-5: 에너지 독립성 (dim=G만) ───────────────────────────────────
P5 = r'에너지\s*독립성'

# ─── C플래그 추가 패턴 ────────────────────────────────────────────────
P_BAT = r'삼성SDI|LG에너지솔루션|SK온'
P_SUP = r'공급망\s*(불안|불안정|충격|차질)|보호무역주의'
P_OWN = r'(최대주주|대주주).{0,20}(변경|교체|취득|매각)'

mask1 = df['matched_sentence'].str.contains(P1, regex=True, na=False)
mask2 = df['matched_sentence'].str.contains(P2, regex=True, na=False)
mask3 = df['matched_sentence'].str.contains(P3, regex=True, na=False)
mask4_all = df['matched_sentence'].str.contains(P4, regex=True, na=False)
mask4_esg = df['matched_sentence'].str.contains('ESG위원회', na=False)
mask4_remove = mask4_all & ~mask4_esg & (df['matched_sentence'].str.len() <= 100)
mask4_cflag  = mask4_all & ~mask4_esg & (df['matched_sentence'].str.len() >  100)
mask5 = df['matched_sentence'].str.contains(P5, regex=True, na=False) & (df['dimension'] == 'G')

mask_remove = mask1 | mask2 | mask3 | mask4_remove | mask5

mask_bat = df['matched_sentence'].str.contains(P_BAT, regex=True, na=False)
mask_sup = df['matched_sentence'].str.contains(P_SUP, regex=True, na=False)
mask_own = df['matched_sentence'].str.contains(P_OWN, regex=True, na=False)

# ─── C플래그 추가 (제거 대상 제외) ─────────────────────────────────
new_cflag = (mask4_cflag | mask_bat | mask_sup | mask_own) & ~mask_remove & ~df['c_stage_flag']
added_cflag = new_cflag.sum()
df.loc[new_cflag, 'c_stage_flag'] = True

print("─── 제거 건수 ──────────────────────────────────────────────")
print(f"  B-1 신주인수권(공백보완): {mask1.sum():>4}건")
print(f"  B-2 전력을 다하:          {mask2.sum():>4}건")
print(f"  B-3 노동집약·노동자본:    {mask3.sum():>4}건")
print(f"  B-4 임원직원목차(≤100자): {mask4_remove.sum():>4}건")
print(f"  B-5 에너지독립성(G):      {mask5.sum():>4}건")
print(f"  합계(중복제거):           {mask_remove.sum():>4}건")
print()
print(f"  C플래그 추가:             {added_cflag:>4}건")

# ─── 제거 실행 ────────────────────────────────────────────────────────
removed_b = mask_remove.sum()
df = df[~mask_remove].reset_index(drop=True)
after_b = len(df)

print()
print(f"B단계 제거: {removed_b}건")
print(f"  전: {before_b:,}  →  후: {after_b:,}")
print(f"C단계 플래그 누계: {df['c_stage_flag'].sum()}건")
print()
print("dimension 분포 (B단계 후):")
print(df['dimension'].value_counts())

df.to_csv("ESG_passages_clean.csv", index=False)
````

````python
sent = pd.read_csv("ESG_passages_clean.csv")

print(f"전체 문장: {len(sent):,}건")
print(f"c_stage_flag=True: {sent['c_stage_flag'].sum()}건")
print()
print("dimension 분포:")
print(sent['dimension'].value_counts())
````

### 분류 기준

| 레이블 | 기준 |
|--------|------|
| **1 (ESG 보존)** | 환경·사회·지배구조 실질 내용 포함 |
| **0 (비ESG 제거)** | 법조문 단순 인용, 재무 절차, 사업 현황 기술 |

| 그룹 | 기본값 | 이유 |
|------|--------|------|
| A2 (상법·정관) | ESG=1 | G 내용 혼재 → 보수적 보존 |
| A3 (작성기준일) | ESG=1 | 이사회 구성 내용 혼재 |
| A4 (재무비율) | ESG=0 | 재무관리 서술 중심 |
| B_배터리 | ESG=0 | 사업 파트너 맥락 |
| B_공급망 | ESG=0 | 시장 현황 서술 |
| B_최대주주 | ESG=0 | 지배구조 변동 사실 |
| B_기타 | ESG=1 | 기타 보수적 보존 |

````python
cflag = sent[sent['c_stage_flag'] == True].copy()

# ─── 그룹 태깅 ───────────────────────────────────────────────
def tag_group(s):
    if re.search(r'상법\s*제\d+조|정관\s*제\d+조|제\d+조\s*제\d+항|제\d+항\s*각\s*호', s):
        return 'A2'
    if re.search(r'작성기준일', s):
        return 'A3'
    if re.search(r'부채비율|유동비율|자기자본비율', s):
        return 'A4'
    if re.search(r'삼성SDI|LG에너지솔루션|SK온', s):
        return 'B_배터리'
    if re.search(r'공급망\s*(불안|불안정|충격|차질)|보호무역주의', s):
        return 'B_공급망'
    if re.search(r'(최대주주|대주주).{0,20}(변경|교체|취득|매각)', s):
        return 'B_최대주주'
    return 'B_기타'

cflag['group'] = cflag['matched_sentence'].apply(lambda x: tag_group(str(x)))
print("그룹 분포:")
print(cflag['group'].value_counts())
````

````python
ESG_POSITIVE = [
    r'독립성.{0,30}(사외이사|감사위원회|이사회|이사)',
    r'(사외이사|감사위원회|이사).{0,30}독립성',
    r'이사의\s*독립성',
    r'사외이사.*자격요건.*충족',
    r'사외이사.*후보.*(검증|추천|독립성)',
    r'사외이사.*과반수.*이사회',
    r'사외이사.*선임.*과정',
    r'감사위원회.*위원.*독립성',
    r'전자투표제.{0,20}(채택|도입|시행)',
    r'(채택|도입).{0,20}전자투표제',
    r'소수주주권', r'주주제안권', r'주주소통', r'주주권익',
    r'주주가치\s*(제고|향상)',
    r'집중투표.{0,20}(도입|삭제|운영|시행)',
    r'의결권.*편의.*제고',
    r'주주환원.*목표|배당.*주주환원',
    r'자기주식\s*소각.{0,30}주주가치|주주가치.{0,30}자기주식\s*소각',
    r'지속가능경영위원회', r'ESG\s*경영',
    r'이사회.{0,20}투명성',
    r'보수위원회.{0,30}(객관적|투명한).{0,30}(심의|심사)',
    r'온실가스\s*(배출량|감축)',
    r'에너지\s*사용량.{0,30}(기재|보고)',
    r'저탄소.{0,20}(기본법|녹색성장)',
    r'(물환경보전법|대기환경보전법|폐기물.{0,10}법).{0,30}(위반|과태료|신고)',
    r'환경관련\s*규제사항',
    r'준법통제기준.{0,30}(준수점검|점검.*결과|활동)',
    r'준법.{0,10}점검\s*활동',
    r'준법지원인.{0,30}(활동내역|점검.*결과|보고)',
    r'준법통제.{0,20}기준.{0,20}(마련|수립)',
]

NON_ESG_NEGATIVE = [
    r'신주.{0,10}(배정|인수권|모집)', r'신주인수.{0,10}청약',
    r'인수인.{0,10}(주선인|수요예측)', r'납입기일.{0,10}(2주|주주에게\s*통지)',
    r'이사회의\s*결의로\s*정한다', r'단수주.{0,20}이사회의\s*결의',
    r'주주명부.{0,10}(폐쇄|기재변경|정지)', r'정관\s*제\d+조.{0,20}주주명부',
    r'소규모\s*합병', r'흡수합병.{0,30}(합병비율|합병기일|합병에\s*관한\s*주주총회)',
    r'합병비율.{0,20}합병기일', r'채무보증.{0,20}이사회\s*결의',
    r'지급보증.{0,20}(이사회|승인)', r'담보제공.{0,20}이사회\s*결의',
    r'인적사항은\s*(아래|다음)와\s*같습니다',
    r'(현황은|주요경력은)\s*(아래|다음)와\s*같습니다',
    r'아래와\s*같습니다\s*$', r'다음과\s*같습니다\s*$',
    r'이사회를\s*소집함에는.{0,30}(개최시기|장소|안건)',
    r'이사회\s*운영규정.{0,30}소집절차',
    r'(부채비율|유동비율).{0,30}(모니터링|관리|기초하여\s*자본)',
    r'자본위험.{0,30}(부채비율|자본구조)',
    r'주주이익의\s*극대화.{0,30}(부채비율|자본구조)',
    r'(삼성SDI|LG에너지솔루션|SK온).{0,50}(납품|공급|점유율|배터리\s*제조|이차전지\s*생산|매출)',
    r'(납품|공급).{0,50}(삼성SDI|LG에너지솔루션|SK온)',
    r'공급망.{0,20}(차질|불안).{0,50}(매출|감소|판매|어려움)',
    r'(지정학적|무역\s*갈등|원자재\s*가격\s*상승).{0,30}공급망',
    r'보호무역주의.{0,50}(판매|매출|성장|경쟁)',
    r'최대주주.{0,20}(변경됨에\s*따라|변경\s*완료|변경.*기한의\s*이익)',
    r'(최대주주|대주주).{0,20}변경.{0,30}(편입|그룹에)',
    r'지배구조변경\s*제한.{0,30}최대주주\s*변경',
    r'(흡수합병|주식교환).{0,30}(상장폐지|합병비율)',
    r'상법\s*제368조의4.{0,50}전자투표.{0,30}행사할\s*수\s*있습니다$',
]

GROUP_DEFAULT = {
    'A2': 1, 'A3': 1, 'A4': 0,
    'B_배터리': 0, 'B_공급망': 0, 'B_최대주주': 0, 'B_기타': 1,
}

POS_PATS = [re.compile(p) for p in ESG_POSITIVE]
NEG_PATS = [re.compile(p) for p in NON_ESG_NEGATIVE]

def classify(sentence, group):
    s = str(sentence)
    if re.search(r'100\s*分\s*의', s):
        return 0, 'hanja'
    for pat in POS_PATS:
        if pat.search(s):
            return 1, f'pos:{pat.pattern[:40]}'
    for pat in NEG_PATS:
        if pat.search(s):
            return 0, f'neg:{pat.pattern[:40]}'
    return GROUP_DEFAULT[group], f'default:{group}'

cflag['esg_label'], cflag['reason'] = zip(
    *cflag.apply(lambda r: classify(r['matched_sentence'], r['group']), axis=1)
)

print("=== 분류 결과 ===")
print(cflag['esg_label'].value_counts())
print()
print("=== 그룹별 ESG=1 비율 ===")
for g in cflag['group'].unique():
    sub = cflag[cflag['group'] == g]
    n1 = (sub['esg_label'] == 1).sum()
    print(f"  {g:15s}: {n1:3d}/{len(sub):3d} 보존")
````

````python
# 전체 문장에 esg_label 반영
sent['esg_label'] = 1
for _, row in cflag.iterrows():
    sent.at[row.name, 'esg_label'] = row['esg_label']

# 한자 分 패턴 추가 제거
hanja_mask = sent['matched_sentence'].str.contains(r'100\s*分\s*의', regex=True, na=False)
sent.loc[hanja_mask, 'esg_label'] = 0
print(f"한자 分 패턴 추가 제거: {hanja_mask.sum()}건")

sent_clean = sent[sent['esg_label'] == 1].drop(columns=['esg_label'])
print(f"총 제거: {(sent['esg_label']==0).sum()}건")
print(f"남은 문장: {len(sent_clean):,}건")

# C_flag_분류결과.csv 저장
cflag[['stock_code', 'fiscal_year', 'section', 'dimension',
       'matched_sentence', 'group', 'esg_label', 'reason']].to_csv(
    "C_flag_분류결과.csv", index=False, encoding='utf-8-sig'
)
print("저장: C_flag_분류결과.csv")

# 문장 레벨 저장
sent_clean.to_csv("ESG_passages_clean_phase2.csv", index=False, encoding='utf-8-sig')
print("저장: ESG_passages_clean_phase2.csv")

# 기업-연도 집계
agg = (
    sent_clean
    .groupby(['stock_code', 'fiscal_year'])['matched_sentence']
    .apply(lambda x: ".. ".join(x))
    .reset_index()
    .rename(columns={'matched_sentence': 'ESG_passage'})
)
agg.to_csv("ESG_passage_기업-연도_final.csv", index=False, encoding='utf-8-sig')
print(f"저장: ESG_passage_기업-연도_final.csv ({len(agg)}행)")
````
