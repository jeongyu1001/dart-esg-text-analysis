## 연구 배경

ESG는 기업가치 평가와 투자 의사결정의 중요한 기준으로 자리 잡고 있으며, 국내에서도 ESG 투자 규모와 책임투자가 빠르게 확대되고 있다. 그러나 실제 시장에서는 ESG를 하나의 점수나 등급으로 이해하는 경향이 강하며, KCGS와 같은 평가기관의 등급이 대표적인 판단 기준으로 활용된다.

하지만 동일한 기업에 대해서도 평가기관마다 상이한 ESG 평가 결과가 나타나고 있으며, 이는 ESG가 단일 지표만으로 설명되기 어려운 복합적인 개념임을 보여준다. 기업의 ESG 수준은 외부 평가기관의 등급뿐 아니라 실제 투자 행동, 그리고 기업이 공시를 통해 전달하는 서술 정보 등 다양한 차원에서 나타날 수 있다.

따라서 ESG를 하나의 점수로 환원하기보다, 서로 다른 정보를 담고 있는 복수의 신호로 이해할 필요가 있다.

## 연구 목적

본 연구는 ESG를 단일 점수가 아닌 **다층적 신호 시스템(Multi-layer ESG Signal System)** 으로 재해석하고자 한다. 이를 위해 ESG를 다음의 세 가지 신호로 구분한다.

* **구조적 신호(Structural Signal)** : 외부 평가기관의 ESG 등급(KCGS)
* **행동 신호(Behavioral Signal)** : 기업의 실제 투자 활동(CAPEX)
* **서술 신호(Narrative Signal)** : 사업보고서에 나타난 ESG 관련 공시(Text)

본 연구는 각 신호가 서로 독립적인 설명력을 가지는지, 그리고 이들을 순차적으로 결합할 때 ESG에 대한 설명력과 예측력이 향상되는지를 실증적으로 검증한다.

> **핵심 연구 질문**
>
> "구조적·행동·서술 신호는 각각 독립적인 정보를 제공하는가? 또한 이들 신호를 결합할 때 ESG에 대한 설명력은 점진적으로 향상되는가?"


## 서술 신호(Text)를 왜 3-Layer로 구축하는가?

본 연구는 ESG를 구조적 신호, 행동 신호, 서술 신호의 세 가지 층위로 재해석한다. 이 중 구조적 신호(KCGS)와 행동 신호(CAPEX)는 이미 정형화된 데이터로 존재하지만, 서술 신호(Text)는 기업의 사업보고서에 분산된 비정형 정보로 존재하므로 별도의 구축 과정이 필요하다.

특히 기업의 ESG 공시는 단순히 ESG 관련 단어를 많이 사용하는 것만으로 설명되지 않는다. 동일한 수준의 ESG 언급을 하더라도 어떤 주제를 다루는지, 그리고 ESG를 어떠한 맥락과 의미로 서술하는지에 따라 전달되는 정보가 달라질 수 있다.

이에 본 연구는 사업보고서의 ESG 서술을 다음의 세 가지 정보 층위로 구분하여 서술 신호를 구축한다.

* **Layer 1 (Intensity)** : ESG 관련 표현을 얼마나 많이 사용하는가
* **Layer 2 (Content)** : 어떤 ESG 주제와 이슈를 다루고 있는가
* **Layer 3 (Semantic)** : ESG를 어떠한 맥락과 의미로 설명하고 있는가

즉, 서술 신호 역시 단일 텍스트 지표가 아니라 빈도, 내용, 의미라는 서로 다른 정보를 담고 있는 다층적 신호 체계로 이해할 수 있다. 본 연구는 이러한 3-Layer 접근을 통해 사업보고서 공시가 제공하는 ESG 정보를 보다 정교하게 포착하고, 각 층위가 독립적인 설명력을 제공하는지를 검증한다.

또한 빈도 신호는 강하지만 내용 및 의미 신호가 상대적으로 약한 경우, 기업이 실질적인 ESG 활동보다 과도한 공시를 통해 ESG 이미지를 강조하는 **Cheap-talk 패턴**이 나타날 가능성을 추가적으로 탐색한다.

## 0.4 분석 프레임워크: ESG 3-Signal

본 연구는 ESG 신호를 **세 층위**로 분리한다.

```
┌────────────────────────────────────────────────────────┐
│              ESG 3-Signal Framework                    │
│                                                        │
│  Signal 1 │ 텍스트 서술  │ DART 사업보고서 공시 언어   │
│           │ (본 연구 핵심)│ → 3-Layer 텍스트 피처      │
│  ─────────┼─────────────┼──────────────────────────────┤
│  Signal 2 │ 실물 행동   │ CAPEX 강도 (자산 대비)       │
│  ─────────┼─────────────┼──────────────────────────────┤
│  Signal 3 │ 외부 평가   │ KCGS ESG 등급 (종속변수)     │
└────────────────────────────────────────────────────────┘
```

**텍스트 신호 내부 3-Layer 구조** (본 과제의 핵심 Feature):

| Layer | 측정 방식 | 변수 | PART |
|---|---|---|---|
| **Intensity** (키워드 빈도) | Seed 사전 기반 TF-IDF | `E/S/G_seed_norm_log` | 4 |
| **Content** (TF-IDF 주제) | SVD 축약 후 유의 PC | `neg_TFPC2`, `TFPC1` | 5 |
| **Semantic** (문장 의미) | KR-FinBert 임베딩 PCA | `neg_KRFPC1`, `KRFPC2` | 6 |

> 세 Layer를 순차적으로 추가할수록 R²가 상승한다 (M1=0.159 → M3b=0.305).  
> 이것이 3-Layer 접근의 핵심 근거다.
