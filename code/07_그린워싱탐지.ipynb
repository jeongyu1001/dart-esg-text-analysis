## 12.1 TriSignal GW 모델 설계 및 근거

## 3. TriSignal 그린워싱 탐지 모델

### 3-1. 핵심 가정 및 근거

본 연구는 위 세 그림의 예비분석 결과를 바탕으로 아래 **세 가지 가정**을 도출하고,
이를 그린워싱 탐지 공식에 반영하였다.

---

**가정 1. 텍스트 서술이 많을수록 그린워싱 의심이 높다** → L1 가중치 **+0.35**

- **[그림 1 (논문 참고)]** 서술(ESG 텍스트) 신호는 상단 꼬리가 두꺼운 우편향 분포를 보인다.  
  일부 기업이 공시량을 극단적으로 늘리는 반면, 행동·평가 신호는 그 폭이 제한적이다.
- **[그림 3(논문 참고)]** 서술↔행동 r=−0.13, 서술↔KCGS r=−0.13으로 음의 상관관계가 확인된다.  
  텍스트 서술이 많은 기업일수록 실물 투자(CAPEX)와 외부 평가(KCGS)가 오히려 낮은 경향이 있다.  
  *(여기서 "서술"은 `ESG_seed_norm`(원시 seed 키워드 밀도) 기준이며, 아래 GW 공식의 L1은 Seed·Content·Semantic을 결합한 `Layer1_text`를 사용한다. `Layer1_text`는 공시 품질까지 반영하므로 KCGS와 양의 상관(r≈+0.39)을 가지나, GW 탐지는 단독 상관이 아니라 L1↑ + L2↓의 교차조건으로 판별하기 때문에 모델 설계에 영향이 없다.)*
- → **ESG 공시량이 많다는 사실 자체가 그린워싱 의심 신호**로 기능할 수 있다.

---

**가정 2. 실제 투자 행동(CAPEX)이 낮을수록 그린워싱 위험이 높다** → L2 가중치 **−0.45**

- **[그림 2 (논문 참고)]** ESG 텍스트 길이(X축)와 로그 CAPEX(Y축)를 교차한 결과,  
  우하단 음영 구역(텍스트↑ + CAPEX↓)에 다수의 기업-연도가 분포한다.  
  이를 **Cheap-talk 가능성** 구역으로 정의한다.
- 추가로, 행동↔KCGS r=+0.65로 실물 투자와 외부 평가는 비교적 동조하는 반면,  
  서술은 이 두 지표 모두와 음의 관계를 보인다. 즉 CAPEX 부재가 곧 실행 결여의 증거다.
- → **공시는 많지만 실물 투자가 낮은 기업 = 핵심 탐지 대상**이다.

---

**가정 3. 같은 텍스트·CAPEX 조건에서 KCGS 등급이 높을수록 그린워싱 위험이 높다** → L3 실효 가중치 **+0.20**

- **[그림 4]** Cheap-talk 구역(텍스트↑ + CAPEX↓)에 **A+, A, B+ 등 고등급 기업**이 포함되어 있다.  
  즉, 서술·행동 조건이 동일할 때 외부 평가만 높은 기업이 실재한다.
- **[그림 5]** 서술(`ESG_seed_norm`)↔KCGS r=−0.13으로, 원시 키워드 공시량이 많은 기업이 반드시 고등급은 아니며  
  역으로 고등급 기업이 텍스트 공시가 적은 경우도 존재한다.  
  이는 KCGS 등급이 단순 서술량과 독립적으로 움직임을 시사한다.
- → **같은 L1·L2 수준에서 외부 등급이 높은 기업은 "평가기관까지 속인" 가능성이 있다**  
  (grader-fooling 해석). 이를 가장 위험한 그린워셔로 간주하여 GW↑ 방향으로 반영하였다.

---

### 3-2. 산업별 GW 점수 산출 공식

텍스트 기반 ESG 서술(L1), 행동 및 산업 맥락(L2), 외부 평가(KCGS, L3)를 결합하여  
산업 특성에 따라 다음과 같이 그린워싱 점수(GW)를 산출하였다.

```
[비금융 + CAPEX 있음]  GW = 0.35×L1 − 0.45×L2 − 0.20×L3
[비금융 + CAPEX 없음]  GW = 0.575×L1            − 0.425×L3
[금융업]               GW = 0.60×L1 − 0.20×L2  − 0.20×L3
```

| 레이어 | 신호 변수 | 공식 계수 | 실효 방향 | 의미 |
|--------|-----------|-----------|-----------|------|
| **L1 (서술)** | Text Seed Score (Z-score) | **+0.35** | 공시 多 → GW↑ | ESG 서술량이 많을수록 그린워싱 의심 상승 |
| **L2 (행동)** | CAPEX 강도 (Z-score) | **−0.45** | 실투자 多 → GW↓ | 실물 행동이 뒤따를수록 그린워싱 위험 감소 |
| **L3 (외부평가)** | KCGS 등급 (−esg_z 적용) | −0.20 → **실효 +0.20** | 고등급 → GW↑ | 텍스트·CAPEX 통제 후 고등급 = 기관 기만 신호 |

> **L3 구현 방식**: `Layer3_validation = −esg_numeric_z` 로 부호를 반전시킨 후 공식 계수 −0.20을 곱하면  
> 실효 가중치는 **+0.20**이 된다. 즉, KCGS 등급이 높을수록(같은 L1·L2 조건 기준) GW 점수가 상승하도록 설계하였다.

### 3-3. 가중치 설정 근거

가중치는 앞선 가정을 기반으로 산업 특성을 고려하여 설정하였다.

- **비금융 기업**: 실제 ESG 실행을 반영하는 행동 신호(L2)의 비중을 높게 부여하였고(−0.45),  
  CAPEX 데이터가 없는 경우 L1·L3만으로 대체 산출한다.
- **금융업**: CAPEX의 ESG 실행 대리지표로서의 해석 한계를 고려해  
  텍스트 기반 서술(L1)의 비중을 확대하였다(0.60).
- **가중치 안정성 검토**: 다양한 가중치 조합 검토 결과, 위험 기업군이 전반적으로 유사하게 도출되어  
  최종 가중치는 **해석 가능성과 산업 특성**을 중심으로 확정하였다.

### 3-4. GW 리스크 분류 기준

GW 점수는 상대적 공시 불일치 수준을 구분하기 위해 **표본 분포 기반 백분위 기준**을 적용하였다.  
절대적 임계값이 아닌 분포 기반 상대 분류를 사용함으로써, 표본 내 **상대적 그린워싱 위험도**를 비교 가능하게 하였다.

| 등급 | 기준 | 의미 |
|------|------|------|
| **High Risk** | 상위 20% (p80 초과) | 서술↑ 행동↓ 패턴이 뚜렷한 고위험군 |
| **Medium Risk** | 20~50% (p50~p80) | 부분적 불일치 존재 |
| **Low Risk** | 하위 50% (p50 이하) | 서술·행동·평가가 비교적 정합 |

## 4. 연구 기여 및 한계

**기여**
1. **다층 신호 체계 재구성**: ESG를 외부평가·행동·서술의 세 층위로 분리하여 각 신호의 독립적 설명력 검증
2. **DART 비정형 텍스트 활용**: ESG 서술과 실제 행동(CAPEX) 간 괴리 가능성을 실증적으로 탐색
3. **복합 신호 기반 접근 필요성 제시**: 단일 ESG 등급이 설명하지 못하는 정보 차이를 확인함으로써  
   ESG 평가의 복합 신호 기반 접근 필요성 제시

**한계 및 분석 범위**
- 본 연구의 분석 중심은 DART 사업보고서의 **비정형 텍스트 기반 ESG 서술 정보의 정량화**에 있으며,  
  외부평가(KCGS)와 행동(CAPEX) 변수는 기존 정량 지표를 **보완적으로** 활용하였다.
- 그린워싱 탐지 모델은 탐색적 확장 분석으로, 가중치는 이론적 가정에 기반하며  
  **외부 검증 데이터(실제 그린워싱 사례)와의 정합성 확인**이 향후 과제로 남는다.
- 이러한 점에서 본 연구는 DART 사업보고서 비정형 텍스트를 활용하여  
  기업의 ESG 서술 정보를 정량화하고 이를 행동 및 평가 데이터와 통합 분석하였다는 점에서  
  **기존 ESG 등급 중심 연구와 차별성**을 가진다.


````python
import pandas as pd
import numpy as np
from scipy.stats import zscore
# ESG 통합 데이터
df = pd.read_csv(
    "data/4차_ESG_integrated_dataset.csv"
)

fin_df = pd.read_csv("fin_df.csv")

# CAPEX 데이터
capex_df = pd.read_csv(
    "capex_df.csv"
)

# 비금융 기업 데이터
non_fin_df = pd.read_csv(
    "financial_data_nonfinancial.csv"
)

kcgs_df = pd.read_csv("kcgs_grade.csv")
````

````python
# -------------------------
# 1. stock_code 형식 맞추기
# -------------------------
for temp_df in [
    df,
    capex_df,
    non_fin_df,
    kcgs_df
]:
    temp_df["stock_code"] = (
        temp_df["stock_code"]
        .astype(str)
        .str.replace(".0", "", regex=False)
        .str.zfill(6)
    )

# -------------------------
# 2. 연도 형식 맞추기
# -------------------------
for temp_df in [
    df,
    capex_df,
    kcgs_df
]:
    temp_df["fiscal_year"] = (
        temp_df["fiscal_year"]
        .astype(int)
    )

# -------------------------
# 3. KCGS 있는 기업만 남기기
# -------------------------
df = df.merge(
    kcgs_df[
        [
            "stock_code",
            "fiscal_year",
            "esg_grade_num"
        ]
    ],
    on=["stock_code", "fiscal_year"],
    how="inner"
)

# -------------------------
# 4. CAPEX merge
# -------------------------
df = df.merge(
    capex_df[
        [
            "stock_code",
            "fiscal_year",
            "CAPEX"
        ]
    ],
    on=["stock_code", "fiscal_year"],
    how="left"
)

# -------------------------
# 확인
# -------------------------
print(df.shape)
df.head()
````

````python
non_fin_codes = set(
    non_fin_df["stock_code"]
)

df["is_financial"] = ~df[
    "stock_code"
].isin(non_fin_codes)
df["is_financial"].value_counts()
````

````python
from scipy.stats import zscore
import numpy as np

# -------------------------
# 1. 사용할 텍스트 변수
# -------------------------
seed_cols = [
    "E_seed_norm_log",
    "S_seed_norm_log",
    "G_seed_norm_log"
]

content_cols = [
    "neg_TFPC2",
    "TFPC1"
]

semantic_cols = [
    "neg_KRFPC1",
    "KRFPC2"
]

all_text_cols = (
    seed_cols +
    content_cols +
    semantic_cols
)

# -------------------------
# 2. z-score 표준화
# -------------------------
for col in all_text_cols:
    df[f"{col}_z"] = zscore(
        df[col].astype(float),
        nan_policy="omit"
    )

# -------------------------
# 3. Seed score (40%)
# -------------------------
seed_z_cols = [
    f"{col}_z"
    for col in seed_cols
]

df["Seed_score"] = (
    df[seed_z_cols]
    .mean(axis=1)
)

# -------------------------
# 4. Content score (30%)
# -------------------------
df["Content_score"] = (
      0.7 * df["neg_TFPC2_z"]
    + 0.3 * df["TFPC1_z"]
)

# -------------------------
# 5. Semantic score (30%)
# -------------------------
df["Semantic_score"] = (
      0.7 * df["neg_KRFPC1_z"]
    + 0.3 * df["KRFPC2_z"]
)

# -------------------------
# 6. 최종 Layer1
# -------------------------
df["Layer1_text"] = (
      0.4 * df["Seed_score"]
    + 0.3 * df["Content_score"]
    + 0.3 * df["Semantic_score"]
)

# 확인
df[
    [
        "Seed_score",
        "Content_score",
        "Semantic_score",
        "Layer1_text"
    ]
].head()
````

## 12.2 L2 행동 신호: CAPEX 강도 산출

````python
import numpy as np
from scipy.stats import zscore

# -------------------------
# 1. CAPEX 강도 만들기
# -------------------------

# 혹시 문자열이면 숫자로
df["CAPEX"] = pd.to_numeric(df["CAPEX"], errors="coerce")
df["assets"] = pd.to_numeric(fin_df["assets"], errors="coerce")

# 자산 대비 CAPEX 비율
df["capex_intensity"] = (
    df["CAPEX"] / df["assets"]
)

# extreme value 완화
df["capex_intensity_log"] = np.log1p(
    df["capex_intensity"]
)
# -------------------------
# 2. 산업 특성 반영하기
# ------------------------- 
action_cols = [
    "capex_intensity_log",
    "E_industry_z"
]

for col in action_cols:
    df[f"{col}_z"] = zscore(
        df[col].astype(float),
        nan_policy="omit"
    )
    
df["Layer2_action"] = np.nan
````

### 12.2.1 비금융 + CAPEX 있음 — GW 공식 적용

````python
non_fin_mask = (
    df["is_financial"] == False
)

df.loc[
    non_fin_mask,
    "Layer2_action"
] = (
    df.loc[
        non_fin_mask,
        [
            "capex_intensity_log",
            "E_industry_z"
        ]
    ].mean(axis=1)
)
````

### 12.2.2 금융업 + CAPEX 없는 경우 — 대체 공식

````python
fin_mask = (
    df["is_financial"] == True
)

df.loc[
    fin_mask,
    "Layer2_action"
] = (
    df.loc[
        fin_mask,
        [
            "E_industry_z"
        ]
    ].mean(axis=1)
)
````

## 12.3 L3 외부평가 신호: KCGS 등급 (−부호 적용)

````python
from scipy.stats import zscore

# Layer3
df["esg_numeric_z"] = zscore(
    df["esg_grade_num"].astype(float),
    nan_policy="omit"
)

# 같은 텍스트·CAPEX 조건에서 등급이 높을수록 GW↑ (평가기관을 속인 신호)
# 공식: -0.20 × (-esg_z) = +0.20 × esg_z → 고등급 → GW 점수 상승
df["Layer3_validation"] = -df["esg_numeric_z"]
````

## 12.4 최종 GW 점수 산출 및 위험도 분류

````python
import numpy as np

# -----------------------------
# 마스크 정의
# -----------------------------
fin_mask = (
    df["is_financial"] == True
)

non_fin_mask = (
    df["is_financial"] == False
)

capex_mask = (
    df["CAPEX"].notna()
)

non_capex_mask = (
    df["CAPEX"].isna()
)

# -----------------------------
# 최종 score 생성
# -----------------------------
df["greenwashing_score"] = np.nan

# =================================
# 1. 비금융 + CAPEX 있음
# =================================
mask1 = (
    non_fin_mask
    & capex_mask
)

df.loc[
    mask1,
    "greenwashing_score"
] = (
    0.35
    * df.loc[mask1, "Layer1_text"]

    - 0.45
    * df.loc[mask1, "Layer2_action"]

    - 0.2
    * df.loc[mask1, "Layer3_validation"]
)

# =================================
# 2. 비금융 + CAPEX 없음
# =================================
mask2 = (
    non_fin_mask
    & non_capex_mask
)

df.loc[
    mask2,
    "greenwashing_score"
] = (
    0.575
    * df.loc[mask2, "Layer1_text"]
    - 0.425
    * df.loc[mask2, "Layer3_validation"]
)

# =================================
# 3. 금융기업
# =================================
mask3 = fin_mask

df.loc[
    mask3,
    "greenwashing_score"
] = (
    0.6
    * df.loc[mask3, "Layer1_text"]

    - 0.2
    * df.loc[mask3, "Layer2_action"]

    - 0.2
    * df.loc[mask3, "Layer3_validation"]
)

# -----------------------------
# 위험도 분류
# -----------------------------
high_cut = (
    df["greenwashing_score"]
    .quantile(0.80)
)

mid_cut = (
    df["greenwashing_score"]
    .quantile(0.50)
)

df["GW_risk"] = np.select(
    [
        df["greenwashing_score"]
        >= high_cut,

        df["greenwashing_score"]
        >= mid_cut
    ],
    [
        "High Risk",
        "Medium Risk"
    ],
    default="Low Risk"
)

# -----------------------------
# 결과 확인
# -----------------------------
df_final = (
    df[
        [
            "company_name",
            "stock_code",
            "fiscal_year",
            "CAPEX",
            "is_financial",
            "greenwashing_score",
            "GW_risk"
        ]
    ]
    .sort_values(
        "greenwashing_score",
        ascending=False
    )
)

df_final.head(30)
````

````python
# -----------------------------
# 위험도 개수
# -----------------------------
risk_count = (
    df["GW_risk"]
    .value_counts()
)

print(risk_count)

# -----------------------------
# 위험도 비율(%)
# -----------------------------
risk_ratio = (
    df["GW_risk"]
    .value_counts(normalize=True)
    * 100
).round(2)

print("\n비율(%)")
print(risk_ratio)
````

### 12.4.1 위험도 분류 기준 (백분위 기반)

| 등급 | 기준 | 의미 |
|------|------|------|
| **High Risk** | 상위 20% (p80 초과) | 서술↑ 행동↓ 패턴이 뚜렷한 고위험군 |
| **Medium Risk** | 20~50% (p50~p80) | 부분적 불일치 존재 |
| **Low Risk** | 하위 50% (p50 이하) | 서술·행동·평가가 비교적 정합 |

````python
import pandas as pd
import numpy as np

# 원본 유지
df_test = df.copy()

weights = [
    (0.5, 0.3, 0.2),
    (0.4, 0.4, 0.2),
    (0.5, 0.2, 0.3),
    (0.4, 0.3, 0.3),
]

top_sets = {}

for w1, w2, w3 in weights:

    df_test["score_test"] = np.nan

    # 마스크
    fin_mask = (
        df_test["is_financial"] == True
    )

    non_fin_mask = (
        df_test["is_financial"] == False
    )

    capex_mask = (
        df_test["CAPEX"].notna()
    )

    non_capex_mask = (
        df_test["CAPEX"].isna()
    )

    # -----------------------------
    # 비금융 + CAPEX 있음
    # -----------------------------
    mask1 = (
        non_fin_mask
        & capex_mask
    )

    df_test.loc[
        mask1,
        "score_test"
    ] = (
        w1
        * df_test.loc[
            mask1,
            "Layer1_text"
        ]
        - w2
        * df_test.loc[
            mask1,
            "Layer2_action"
        ]
        - w3
        * df_test.loc[
            mask1,
            "Layer3_validation"
        ]
    )

    # -----------------------------
    # 비금융 + CAPEX 없음
    # -----------------------------
    mask2 = (
        non_fin_mask
        & non_capex_mask
    )

    # 행동 레이어 제거 후 비율 재분배
    total = w1 + w3

    text_weight = w1 / total
    kcgs_weight = w3 / total

    df_test.loc[
        mask2,
        "score_test"
    ] = (
        text_weight
        * df_test.loc[
            mask2,
            "Layer1_text"
        ]
        - kcgs_weight
        * df_test.loc[
            mask2,
            "Layer3_validation"
        ]
    )

    # -----------------------------
    # 금융기업
    # -----------------------------
    df_test.loc[
        fin_mask,
        "score_test"
    ] = (
        (w1 + 0.1)
        * df_test.loc[
            fin_mask,
            "Layer1_text"
        ]
        - max(w2 - 0.1, 0)
        * df_test.loc[
            fin_mask,
            "Layer2_action"
        ]
        - w3
        * df_test.loc[
            fin_mask,
            "Layer3_validation"
        ]
    )

    # Top30
    top = (
        df_test
        .sort_values(
            "score_test",
            ascending=False
        )
        .drop_duplicates(
            subset="stock_code"
        )
        .head(30)
    )

    companies = (
        top["company_name"]
        .tolist()
    )

    key = f"{w1}-{w2}-{w3}"
    top_sets[key] = set(companies)

    print(f"\n가중치 {key}")
    print(companies[:10])


# -----------------------------
# 공통 기업 수 계산
# -----------------------------
common = set.intersection(
    *top_sets.values()
)

print("\n공통 기업 수:", len(common))
print(sorted(common))
````
