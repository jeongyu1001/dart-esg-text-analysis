## 6.1 KR-FinBert 임베딩 선택 배경

**왜**: KoSimCSE(범용 한국어)보다 KR-FinBert(금융 도메인 특화)가 사업보고서의 ESG 문맥을 더 잘 포착한다.  
**방법**: STEP3-1 문장 분할(`.` 기준) → STEP3-2 KR-FinBert 768차원 임베딩 → STEP3-3 문서 평균 풀링.  
**결정사항**: KR-FinBert 채택 — OLS M3b R²=0.298 > KoSimCSE M3a R²=0.286 (+0.012).

```python
# 필요 패키지 설치 (최초 1회)
# !pip install transformers torch sentencepiece
```

```python
import sys, os
sys.path.insert(0, os.path.abspath('.'))
import config
import pandas as pd
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModel
import warnings
warnings.filterwarnings('ignore')

# ── 데이터 로드 ──────────────────────────────────────────
df = config.load('passage_final')
print(f'입력 데이터: {df.shape}')
print(f'컬럼: {list(df.columns)}')
```

## 6.1.1 문장 분할 (STEP3-1)

## STEP3-1. 문장 분할 (by '.')

```python
# 문장 분할
records = []
for _, row in df.iterrows():
    sentences = [s.strip() for s in str(row['ESG_passage']).split('..') if len(s.strip()) > 10]
    for sent in sentences:
        records.append({
            'stock_code': row['stock_code'],
            'fiscal_year': row['fiscal_year'],
            'sentence': sent
        })

sent_df = pd.DataFrame(records)
print(f'총 문장 수: {len(sent_df):,}')
print(f'기업-연도 수: {sent_df.groupby(["stock_code","fiscal_year"]).ngroups}')
```

## 6.1.2 KR-FinBert 임베딩 768차원 (STEP3-2)

## STEP3-2. KR-FinBert 임베딩 (768차원)

```python
# ── 모델 로드 ────────────────────────────────────────────
MODEL_NAME = 'snunlp/KR-FinBert'
print(f'모델 로드: {MODEL_NAME}')

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model     = AutoModel.from_pretrained(MODEL_NAME)
model.eval()

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model  = model.to(device)
print(f'디바이스: {device}')
```

```python
def get_embedding(texts, batch_size=32):
    """텍스트 리스트 → 문장별 768차원 임베딩 (mean pooling)"""
    all_embeddings = []
    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]
        encoded = tokenizer(
            batch,
            padding=True,
            truncation=True,
            max_length=512,
            return_tensors='pt'
        ).to(device)
        with torch.no_grad():
            output = model(**encoded)
        # Mean pooling (attention mask 반영)
        mask = encoded['attention_mask'].unsqueeze(-1).float()
        emb  = (output.last_hidden_state * mask).sum(1) / mask.sum(1)
        all_embeddings.append(emb.cpu().numpy())
        if (i // batch_size) % 10 == 0:
            print(f'  처리 중: {i+len(batch)}/{len(texts)}')
    return np.vstack(all_embeddings)

print('문장 임베딩 생성 중...')
sentences = sent_df['sentence'].tolist()
emb_matrix = get_embedding(sentences)
print(f'임베딩 완료: {emb_matrix.shape}')
```

## 6.1.3 문서 평균 풀링 (STEP3-3)

## STEP3-3. 문서 평균 풀링 (문장 → 기업-연도)

```python
# 문장 임베딩 → 문서(기업-연도) 평균 풀링
sent_df[[f'emb_{i}' for i in range(768)]] = emb_matrix

emb_cols = [f'emb_{i}' for i in range(768)]
doc_emb  = (
    sent_df
    .groupby(['stock_code','fiscal_year'])[emb_cols]
    .mean()
    .reset_index()
)
# 컬럼명 통일 (KoSimCSE와 동일 형식)
doc_emb.columns = ['stock_code','fiscal_year'] + [f'embedding_{i}' for i in range(768)]

print(f'문서 임베딩: {doc_emb.shape}')

out_path = 'ESG_KRFinBert_doc_embedding (KR-FinBert방식).csv'
doc_emb.to_csv(out_path, index=False, encoding='utf-8-sig')
print(f'저장: {out_path}')
```

## 6.2.1 PCA 적용 (누적 분산 80% 기준)

**왜**: 768차원 임베딩을 그대로 회귀변수로 쓰면 다중공선성·과적합 위험이 있다.  
**방법**: PCA; 누적 설명 분산 80% 이상을 기준으로 최적 성분 수 결정.  
**결정사항**: Y 독립적 기준(분산 상위 2개: KRFPC1·KRFPC2) 채택 — ESG 상관은 확인용으로만 사용.

```python
import sys, os
sys.path.insert(0, os.path.abspath('.'))
import config
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# ── 데이터 로드 ──────────────────────────────────────────
emb_path = '../STEP3-1_3_KoSimCSE임베딩/ESG_KRFinBert_doc_embedding (KR-FinBert방식).csv'
emb = pd.read_csv(emb_path)
emb['stock_code'] = emb['stock_code'].astype(str).str.zfill(6)

grade = config.load('integrated')
grade['stock_code'] = grade['stock_code'].astype(str).str.zfill(6)

emb_cols = [c for c in emb.columns if c.startswith('embedding_')]
X = emb[emb_cols].values
print(f'임베딩: {X.shape}')
```

## STEP3-4. PCA 적용 (누적 80% 기준)

```python
# ── 최적 성분 수 결정 ─────────────────────────────────────
pca_full = PCA(random_state=42)
pca_full.fit(X)
cumvar = np.cumsum(pca_full.explained_variance_ratio_)
n_80   = int(np.argmax(cumvar >= 0.80)) + 1
print(f'누적 80% 기준 성분 수: {n_80}')

# ── PCA 실행 ─────────────────────────────────────────────
pca = PCA(n_components=n_80, random_state=42)
scores = pca.fit_transform(X)
print(f'설명분산: {pca.explained_variance_ratio_.sum()*100:.1f}%')

# 성분 데이터프레임
pc_names = [f'KRFPC{i+1}' for i in range(n_80)]
pca_df = emb[['stock_code','fiscal_year']].copy()
for i, name in enumerate(pc_names):
    pca_df[name] = scores[:, i]

# 설명분산 저장
var_df = pd.DataFrame({
    'component': pc_names,
    'explained_variance': pca.explained_variance_ratio_,
    'cumulative_variance': cumvar[:n_80]
})
var_df.to_csv('krfinbert_pca_variance (KR-FinBert방식).csv', index=False, encoding='utf-8-sig')
print('저장: krfinbert_pca_variance (KR-FinBert방식).csv')
```

## 6.2.2 PC-ESG 상관분석 (STEP3-5)

## STEP3-5. PC-ESG 상관분석

### PC 선택 방침
- **Y 독립적 기준**: 분산 기준 상위 2개 (KRFPC1, KRFPC2)
- KoSimCSE의 SEPC1(분산 27.8%) + SEPC7(유일 반대방향) 선택 방식과 동일 원칙 적용
- ESG 상관은 '확인용'으로만 사용 (선택 기준 아님)

```python
# ── PC × ESG 상관 ─────────────────────────────────────────
combined = pca_df.merge(
    grade[['stock_code','fiscal_year','esg_numeric','e_numeric','g_numeric','s_numeric']],
    on=['stock_code','fiscal_year'], how='left'
)

rows = []
for pc in pc_names:
    for dim, col in [('E','e_numeric'),('G','g_numeric'),('S','s_numeric')]:
        r_p, p_p = stats.pearsonr(combined[pc], combined[col])
        r_s, p_s = stats.spearmanr(combined[pc], combined[col])
        rows.append({'PC':pc,'dimension':dim,
                     'pearson_r':round(r_p,4),'pearson_p':round(p_p,4),
                     'spearman_r':round(r_s,4),'spearman_p':round(p_s,4)})

corr_df = pd.DataFrame(rows)
corr_df.to_csv('PC_ESG_correlation (KR-FinBert방식).csv', index=False, encoding='utf-8-sig')

# 상위 PC 상관 출력
print('=== 상위 10개 PC × ESG 상관 (|Pearson r| 기준) ===')
top = corr_df.copy()
top['abs_r'] = top['pearson_r'].abs()
print(top.nlargest(15,'abs_r')[['PC','dimension','pearson_r','pearson_p']].to_string(index=False))
```

```python
# ── 최종 채택 PC: Y 독립적 기준 ───────────────────────────
# KRFPC1: 가장 큰 성분 (분산 1위)
# KRFPC_?: 방향이 반전되는 첫 번째 PC (KoSimCSE의 SEPC7 역할)

# 방향 반전 PC 탐색 (KRFPC1과 ESG 상관 방향이 다른 PC)
esg_corr = {}
for pc in pc_names[:10]:  # 상위 10개에서만 탐색
    r, _ = stats.pearsonr(combined[pc], combined['esg_numeric'])
    esg_corr[pc] = r

pc1_dir = np.sign(esg_corr['KRFPC1'])
opp_pcs = [pc for pc, r in esg_corr.items() if np.sign(r) != pc1_dir and abs(r) > 0.1]

print('=== KRFPC1 방향:', 'positive' if pc1_dir > 0 else 'negative')
print('=== 방향 반전 후보 PC:', opp_pcs)
print()

adopted_pc2 = opp_pcs[0] if opp_pcs else 'KRFPC2'
print(f'채택 PC: KRFPC1 + {adopted_pc2}')
print(f'  KRFPC1 × ESG: r={esg_corr["KRFPC1"]:+.3f}')
if adopted_pc2 in esg_corr:
    print(f'  {adopted_pc2} × ESG: r={esg_corr[adopted_pc2]:+.3f}')

# neg 방향 처리
r1, _ = stats.pearsonr(combined['KRFPC1'], combined['esg_numeric'])
if r1 < 0:
    pca_df['neg_KRFPC1'] = -pca_df['KRFPC1']
    print('→ neg_KRFPC1 추가 (방향 반전)')
    krfpc1_col = 'neg_KRFPC1'
else:
    krfpc1_col = 'KRFPC1'
```

```python
# ── 최종 저장 ─────────────────────────────────────────────
sig_cols = ['stock_code','fiscal_year', krfpc1_col, adopted_pc2]
if 'neg_KRFPC1' in pca_df.columns and krfpc1_col == 'neg_KRFPC1':
    pca_df['neg_KRFPC1'] = -pca_df['KRFPC1']

sig_df = pca_df[sig_cols]
sig_df.to_csv('krfinbert_sig_pcs (KR-FinBert방식).csv', index=False, encoding='utf-8-sig')
print(f'저장: krfinbert_sig_pcs (KR-FinBert방식).csv  {sig_df.shape}')

# 전체 PCA 저장
pca_df.to_csv('krfinbert_pca_embedding (KR-FinBert방식).csv', index=False, encoding='utf-8-sig')
print(f'저장: krfinbert_pca_embedding (KR-FinBert방식).csv  {pca_df.shape}')
```
